```python
# Mapping for 'Category' to numerical values

Brand_map = {'Apple': 1, 'Sony': 2, 'Samsung': 3}
Sales_2023_new_df['Brand_Numeric'] = Sales_2023_new_df['Product_Brand'].map(Brand_map)

# Display the updated DataFrame
print(Sales_2023_new_df)
```

```python
correlationBrandAge = Sales_2023_new_df['Age'].corr(Sales_2023_new_df['Brand_Numeric'])
print(f"The correlation between Brand and Age: {correlationAge:.2f}")
```

```python
#linear regression model

slope, intercept, r_value, p_value, std_err = stats.linregress(Sales_2023_new_df['Age'], Sales_2023_new_df['Brand_Numeric'])


#to create the plot
x_values = Sales_2023_new_df['Age']
y_values = slope * x_values + intercept
line_eq= "y ="+ str(round(slope,2)) + "x+ " + str(round(intercept,2))

plt.figure(figsize=(12, 6))
plt.scatter(Sales_2023_new_df['Age'], Sales_2023_new_df['Brand_Numeric'], color='blue')
plt.plot(x_values, y_values, color='red')
plt.annotate(line_eq,(5.8,0.8),fontsize=15,color="red")
plt.title(f"The correlation between Brand and Age: {correlationBrandAge:.2f}")
plt.xlabel('# Age')
plt.ylabel('Brand')
print(f"The r-squared is: {r_value**2}")
print(f"Slope: {slope}")
print(f"Intercept: {intercept}")
print(f"P-value: {p_value}")
print(f"Standard Error: {std_err}")

plt.show()

```
